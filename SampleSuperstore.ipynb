# Visualizing the Dataset.

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(),annot=True)      #Heatmap
```

# Converting df['Profit'] to a categorical Data

```python
prof_loss=[]

for i in df['Profit'].values:
    positive='Profit'
    negative="Loss"
    if i >= 0:
        prof_loss.append(positive)  
    else:
        prof_loss.append(negative)
```

```python
df['Profit-Loss']=prof_loss
df['Profit-Loss']
df.head()
```

```python
plt.figure(figsize=(15,15))
sns.pairplot(df,hue='Profit-Loss') #plot Pairwise relationship between the dataset. 
```

From the above pairplot If the discount is high profit will be reduced.

# Visualizing using Piechart

# Pie chart for Ship Mode

```python
plt.figure(figsize=(8,8))
plt.pie(df['Ship Mode'].value_counts(),labels=df['Ship Mode'].value_counts().index,autopct='%.2f%%')
plt.title('SHIP MODE')
plt.show()
```

From the above pieChart we can conclude that Standard class is mostly used.

# Piechart For Segment. 

```python
plt.figure(figsize=(8,8))
plt.pie(df['Segment'].value_counts(),labels=df['Segment'].value_counts().index,autopct='%.2f%%')
plt.title('SEGMENT')
plt.show()
```

From the above chart Segment has High percent.

# Pie Chart for Category

```python
plt.figure(figsize=(8,8))
plt.pie(df['Category'].value_counts(),labels=df['Category'].value_counts().index,autopct='%.2f%%')
plt.title('CATEGORY')
plt.show()
```

From the above piechart we can conclude that Office Supplies are  highly used in Category
